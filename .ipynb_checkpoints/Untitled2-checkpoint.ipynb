This table represents our data including lncons (cigarette consumption as the quantity demanded for cigarettes), lnprice (price for cigarettes), lntax (tax rates for cigarettes), lntipi (the Total Industry Product Index (TIPI)that is highly correlated with the Gross Domestic Product (GDP) as income), and dummy variables (D1, D2, D3, D4, and D5) respresenting the anti-tobacco measures as regulations and tax rate changes.  

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
from matplotlib import rcParams
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
data.describe()
```

```python
plt.scatter(data.lntax, data.lnprice)
plt.xlabel("Tax Rate (lntax)")
plt.ylabel("Cigarette Price (lnprice)")
plt.title("Relationship between Tax and Price")
```

```python
sns.regplot(y="lnprice", x="lntax", data=data, fit_reg = True)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(data.lntax, data.lnprice)
print(corr)
```

First of all, I checked the relationhsip between tax rates and cigarette prices. Both scatter plots and correlation coefficients suggest that there is a strong correlation and/or positive relationship between tax rates and prices meaning that when government increases tax rates price for cigarettes also increases.  

```python
plt.scatter(data.lncons, data.lnprice)
plt.xlabel("Consumption (lncons)")
plt.ylabel("Cigarette Price (lnprice)")
plt.title("Relationship between Cigarette Consumption and Price")
```

```python
sns.regplot(y="lnprice", x="lncons", data=data, fit_reg = True)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(data.lncons, data.lnprice)
print(corr)
```

Second, I checked the relationship between demand or consumption for cigarettes and price. Both scatter plots and correlation coefficient suggest that there is a strong and negative relationship between consumption and price for cigarettes meaning that demand for cigarettes decreases when price increases.  

```python
plt.scatter(data.lncons, data.lntax)
plt.xlabel("Consumption (lncons)")
plt.ylabel("Tax (lntax)")
plt.title("Relationship between Cigarette Consumption and Tax")
```

```python
sns.regplot(y="lntax", x="lncons", data=data, fit_reg = True)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(data.lntax, data.lncons)
print(corr)
```

Third, I checked the relationship between tax rates and consumption. Both scatter plots and correlation coefficient suggest that there is a negative relationship between tax rates and consumption meaning that when government increases tax rates there will be a decreas in cigarette consumption, as expected.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
m = ols('lncons ~ lnprice',data).fit()
print(m.summary())
```

```python
m = ols('lncons ~ lntax',data).fit()
print(m.summary())
```

Lastly, I regressed consumption with price and tax rates to see the regression relationships between those variables. Because coefficients have negative signs (-0.4410 and -0.0810) in both models we can infer that there is a negative relationship between consumption and tax/price as expected. Also, because I used all the variables in the logarithmic form we can interpret those coefficients as demand elasticities. Accordingly, demand elasticities are as expected.
